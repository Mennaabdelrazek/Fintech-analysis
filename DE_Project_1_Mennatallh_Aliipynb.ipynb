#missing data


```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# add a missing indicator column for each feature being analyzed
df['emp_title_missing'] = df['emp_title'].isnull()
df['emp_length_missing'] = df['emp_length'].isnull()
df['annual_inc_joint_missing'] = df['annual_inc_joint'] == 0
df['int_rate_missing'] = df['int_rate'].isnull()
df['description_missing'] = df['description'] == 'not provided'
```

```python
# Convert the 'term' column to string type before using .str.extract
df['term'] = df['term'].astype(str).str.extract('(\d+)').astype(float)
# 1. analyze correlations for 'emp_title' and 'emp_length' (mar hypothesis)
correlation_emp_title = df.groupby('emp_title_missing')[['annual_inc']].mean()
correlation_emp_length = df.groupby('emp_length_missing')[['annual_inc']].mean()

# 2. check correlations for 'annual_inc_joint' (mnar hypothesis)
correlation_annual_inc_joint = df.groupby('annual_inc_joint_missing')['type'].value_counts(normalize=True)

# 3. analyze patterns for 'int_rate' (mar hypothesis)
int_rate_corr = df.groupby('int_rate_missing')[['loan_amount', 'grade', 'term']].mean()

# 4. check if 'description' is mcar by looking for randomness
# count the proportion of missing descriptions across categorical variables
description_corr = df.groupby('description_missing')[['purpose']].value_counts(normalize=True) # Remove 'home_ownership'
print("1. correlation with emp_title_missing:\n", correlation_emp_title)
print("\n2. correlation with emp_length_missing:\n", correlation_emp_length)
print("\n3. correlation with annual_inc_joint_missing:\n", correlation_annual_inc_joint)
print("\n4. correlation with int_rate_missing:\n", int_rate_corr)
print("\n5. correlation with description_missing:\n", description_corr)
```

1.emp_title_missing and emp_length_missing

Correlation: Both emp_title_missing and emp_length_missing show a negative correlation with annual_inc. This means that when emp_title or emp_length is missing, the average annual_inc is lower.
 This is likely MAR (Missing at Random) because the missingness is related to another observed variable (annual_inc).


2. annual_inc_joint_missing

Correlation: We see that when annual_inc_joint is missing (annual_inc_joint_missing is True) the loans are predominantly of the "individual" type. This makes sense as joint applications would require this information, whereas individual applications do not.
Implication: The missingness of annual_inc_joint is directly related to the loan type and is therefore likely MNAR (Missing Not at Random). The missingness is related to the missing values themselves.

3. description_missing

Correlation: We can see that the proportions of purposes differ between cases where description is missing.
Implication: This is likely MAR. It implies the missingness is likely dependent on the purpose of the loan application. For example, those applying for debt consolidation loans may be less likely to provide a detailed description.

```python
def calculate_monthly_installment(P, r_annual, n):
    if pd.isnull(r_annual):  # Handle missing interest rates
        return np.nan
    r = (r_annual / 100) / 12  # Convert annual interest rate to monthly
    n_months = int(n)  # Directly convert the numeric term to an integer
    return P * r * ((1 + r) ** n_months) / (((1 + r) ** n_months) - 1)

df_clean['installment_per_month'] = df_clean.apply(
    lambda x: calculate_monthly_installment(x['loan_amount'], x['int_rate'], x['term']), axis=1
)
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder


# 1. Identify categorical features
categorical_features = ['emp_length', 'home_ownership', 'verification_status', 'purpose', 'loan_status',  'type', 'grade']

# 2. Create a OneHotEncoder object
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore') # sparse=False for dense output

# 3. Fit the encoder to the categorical features
encoder.fit(df_clean[categorical_features])

# 4. Transform the categorical features into encoded columns
encoded_data = encoder.transform(df_clean[categorical_features])

# 5. Create column names for the encoded features
encoded_feature_names = encoder.get_feature_names_out(categorical_features)

# 6. Create a DataFrame from the encoded data
encoded_df = pd.DataFrame(encoded_data, columns=encoded_feature_names, index=df_clean.index)

# 7. Concatenate the encoded DataFrame with the original DataFrame
df_encoded = pd.concat([df_clean, encoded_df], axis=1)

# 8. Drop the original categorical features
df_encoded.drop(columns=categorical_features, inplace=True)
```

One-Hot Encoding: We used one-hot encoding because it avoids imposing an ordinal relationship on categorical features where none exists.

```python
#Normalization
# Those feature likely has a wide range of values and might be skewed.
#MinMax Scaling is chosen for this dataset. Here's why:
#It preserves the original distribution of the data while scaling it to a specific range (usually 0 to 1).
#from sklearn.preprocessing import MinMaxScaler
#and the the data is not normally distributed

from sklearn.preprocessing import MinMaxScaler
features_to_normalize = ['loan_amount', 'annual_inc', 'int_rate']
scaler = MinMaxScaler()
df_encoded[features_to_normalize] = scaler.fit_transform(df_encoded[features_to_normalize])
```
